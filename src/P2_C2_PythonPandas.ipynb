#### 203 DataFrame

```python
v = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
i = ['첫째행', '둘째행', '셋째행']
c = ['컬럼1', '칼럼2', '칼럼3']

df = pd.DataFrame(v, index=i, columns=c)
print(df)
```

### (3) DataFrame 생성하기

#### 204 리스트로 생성하기

### (9) DataFrame 데이터 정렬과 집계

#### 228 오름차순 정렬

```python
housing_sorted = housing_df.sort_values(by=['housing_median_age'])
housing_sorted.head(3)
```

#### 229 여러개 컬럼 내림차순 정렬

```python
housing_sorted = housing_df.sort_values(by=['housing_median_age', 'total_rooms'], ascending=False)
housing_sorted.head(3)
```

#### 230 집계 함수 적용

```python
housing_df.sum() / housing_df.count()
```

#### 231 특정 컬럼 집계 함수 적용

```python
housing_df[['housing_median_age', 'total_rooms']].mean()
```

#### 232 groupby 함수 적용

```python
housing_groupby = housing_df.groupby('housing_median_age').mean()
housing_groupby.head(3)
```

#### 233 특정 컬럼 groupby 함수 적용

```python
housing_groupby = housing_df.groupby('housing_median_age')[['total_rooms', 'total_bedrooms']].mean()
housing_groupby.head(3)
```

#### 234 특정 컬럼 groupby 복수개 함수 적용

```python
housing_groupby = housing_df.groupby('housing_median_age')['total_rooms'].agg([min, max, sum])
housing_groupby.head(3)
```

### (10) DataFrame 데이터 결측값 처리하기

#### 235 결측값 여부 확인

```python
import numpy as np

housing_df['Age_na'] = np.nan
housing_df.isna().head(3)
```

#### 236 결측값 개수 확인

```python
housing_df.isna().sum()
```

#### 237 결측값 데이터 대체하기

```python
housing_df['Age_na'] = housing_df['Age_na'].fillna(housing_df['housing_median_age'].mean())
housing_df.head(3)
```
